```python
import numpy as np


# Функция активации: сигмоидная
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


# Производная сигмоиды
def sigmoid_derivative(x):
    return x * (1 - x)


# Входные данные (X) и выходные данные (y)
# Пример: простая логическая операцию AND
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

y = np.array([[0],
              [0],
              [0],
              [1]])

# Установка случайного начального значения генератора
np.random.seed(42)

# Инициализация весов
input_layer_neurons = X.shape[1]  # 2
hidden_layer1_neurons = 3  # 3 нейрона в первом скрытом слое
hidden_layer2_neurons = 3  # 3 нейрона во втором скрытом слое
output_neurons = 1  # 1 выход

# Случайные веса
weights_input_hidden1 = np.random.uniform(size=(input_layer_neurons, hidden_layer1_neurons))
weights_hidden1_hidden2 = np.random.uniform(size=(hidden_layer1_neurons, hidden_layer2_neurons))
weights_hidden2_output = np.random.uniform(size=(hidden_layer2_neurons, output_neurons))

# Обучение нейронной сети
epochs = 10000
learning_rate = 0.1

for epoch in range(epochs):
    # Прямое распространение
    hidden_layer1_activation = np.dot(X, weights_input_hidden1)
    hidden_layer1_output = sigmoid(hidden_layer1_activation)

    hidden_layer2_activation = np.dot(hidden_layer1_output, weights_hidden1_hidden2)
    hidden_layer2_output = sigmoid(hidden_layer2_activation)

    output_layer_activation = np.dot(hidden_layer2_output, weights_hidden2_output)
    predicted_output = sigmoid(output_layer_activation)

    # Вычисление ошибки
    error = y - predicted_output

    # Обратное распространение
    d_predicted_output = error * sigmoid_derivative(predicted_output)
    error_hidden2_layer = d_predicted_output.dot(weights_hidden2_output.T)
    d_hidden2_layer = error_hidden2_layer * sigmoid_derivative(hidden_layer2_output)

    error_hidden1_layer = d_hidden2_layer.dot(weights_hidden1_hidden2.T)
    d_hidden1_layer = error_hidden1_layer * sigmoid_derivative(hidden_layer1_output)

    # Обновление весов
    weights_hidden2_output += hidden_layer2_output.T.dot(d_predicted_output) * learning_rate
    weights_hidden1_hidden2 += hidden_layer1_output.T.dot(d_hidden2_layer) * learning_rate
    weights_input_hidden1 += X.T.dot(d_hidden1_layer) * learning_rate

# Результаты после обучения
print("Предсказанные выходные данные после обучения:")
print(predicted_output)

# Проверка на обучающих данных
for i in range(X.shape[0]):
    print(f"Вход: {X[i]}, Прогноз: {predicted_output[i]}")
```
